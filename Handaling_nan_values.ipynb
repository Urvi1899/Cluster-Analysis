```python
import re #regular expression
import string

def clean_text(text):
    text = re.sub('[$]',' ', text)
    text = re.sub('[%]',' ', text)
    text=re.sub('[,]','',text)
    return text

clean= lambda x: clean_text(x)
```

```python
df['GDP'] = df['GDP'].astype(str)
df['GDP'] = df['GDP'].apply(clean)
df['GDP'] = df['GDP'].astype(float)
```

```python
df['Health Exp/Capita'] = df['Health Exp/Capita'].astype(str)
df['Tourism Inbound'] = df['Tourism Inbound'].astype(str)
df['Tourism Outbound'] = df['Tourism Outbound'].astype(str)
df['Business Tax Rate'] = df['Business Tax Rate'].astype(str)
```
